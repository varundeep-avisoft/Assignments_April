```python
# a function for removing the higly corrlated columns to avoid multicollinearty
def remove_correlated_features(dataset,threshold):
    columns_corr=set()
    corr_matrix=dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if (corr_matrix.iloc[i,j]) > threshold:
                colname=corr_matrix.columns[i]
                columns_corr.add(colname)
    return columns_corr
```

```python
# dropping the highly correlated columns from the dataframe
correlated_columns=remove_correlated_features(dataframe_heart_disease,0.7)
print(correlated_columns)
dataframe_heart_disease.drop(correlated_columns,inplace=True,axis=1)
dataframe_heart_disease
```

```python
# separating the target and independent columns
X=dataframe_heart_disease.drop('TenYearCHD',axis=1)
y=dataframe_heart_disease['TenYearCHD']
```

```python
X
```

```python
# getting the Inter quantile range of the independent columns
Q1=X.quantile(0.25)
Q3=X.quantile(0.75)
IQR=Q3-Q1
print(IQR)
```

```python
# preparing the outlier mask for removing the outliers
outlier_mask=((X < (Q1 - 1.5*IQR)) | (X > (Q3 + 1.5*IQR))).any(axis=1)
print(outlier_mask)
```

```python
# removing the outliers
cleaned_data_X=X[~outlier_mask]
cleaned_data_y=y[~outlier_mask]
```

```python
print(cleaned_data_X.shape)
print(cleaned_data_y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score,classification_report
```

```python
y.value_counts()
```

```python
# balancing the imbalance in the target variable by oversampling using Smote
from imblearn.combine import SMOTETomek
synthetic_sampler=SMOTETomek(random_state=139)
X_resampled,y_resampled=synthetic_sampler.fit_resample(cleaned_data_X,cleaned_data_y)
```

```python
print(X_resampled.shape)
print(y_resampled.shape)
```

```python
#Scaling the data on the same scale using min max scaler
Scaler=MinMaxScaler()
X_resampled_scaled=Scaler.fit_transform(X_resampled)
```

```python

xtrain,xtest,ytrain,ytest=train_test_split(X_resampled_scaled,y_resampled,random_state=42,test_size=0.3)
```

```python
# training the model with logistic regression
model_logistic_regression=LogisticRegression()
model_logistic_regression.fit(xtrain,ytrain)
```

```python
predictions=model_logistic_regression.predict(xtest)
print("accuracy score of the logistic regression model : ",accuracy_score(ytest,predictions))
print("Classification report of the logistic regression model : \n",classification_report(ytest,predictions))
```
