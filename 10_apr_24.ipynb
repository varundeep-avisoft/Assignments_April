```python
# making the quantiles for getting inter quantile range
Q1=dataframe_breast_cancer_after_feature_selection.quantile(0.25)
Q3=dataframe_breast_cancer_after_feature_selection.quantile(0.75)

# getting the Inter quantile range
IQR=Q3-Q1
print(IQR)
```

```python
# getting the outlier mask for removing the outliers from the dataframe
outlier_mask=((dataframe_breast_cancer_after_feature_selection < (Q1-1.5*IQR)) | (dataframe_breast_cancer_after_feature_selection > (Q3+1.5*IQR))).any(axis=1)
print(outlier_mask)
```

```python
# applying the outlier mask for to get cleaned dataset
cleaned_data_Breast_cancer_X=dataframe_breast_cancer_after_feature_selection[~outlier_mask]
cleaned_data_Breast_cancer_y=y[~outlier_mask]
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import MinMaxScaler,LabelEncoder
from sklearn.linear_model import LogisticRegression
```

```python
# scaling the indpendent features between zero and one
scaler=MinMaxScaler()
cleaned_data_Breast_cancer_X=scaler.fit_transform(cleaned_data_Breast_cancer_X)
cleaned_data_Breast_cancer_X
```

```python
# converting the categories of target column into numerical values using label encoding
encoder=LabelEncoder()
cleaned_data_Breast_cancer_y=encoder.fit_transform(cleaned_data_Breast_cancer_y)
cleaned_data_Breast_cancer_y
```

```python
#SPLITTTING DATA FRAME IN TRAINING AND TESTING
xtrain,xtest,ytrain,ytest = train_test_split(cleaned_data_Breast_cancer_X,cleaned_data_Breast_cancer_y,test_size=0.3,random_state=42)
```

```python
model_logistic_regression=LogisticRegression(max_iter=10000)
```

```python
# parameter grid for hyperparameter tuning
param_grid = {
    'penalty': ['l1', 'l2'],
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'solver': ['liblinear', 'saga']
}
```

```python
# hyper parameter tuning
grid_search = GridSearchCV(estimator=model_logistic_regression, param_grid=param_grid, cv=5)

# Fit the grid search to the data
grid_search.fit(xtrain,ytrain)

# Print the best parameters found
print("Best parameters found:")
print(grid_search.best_params_)

# Print the best score found
print("Best score found:")
print(grid_search.best_score_)
```

```python
# traing model with logistic regression with best parameters
model_logistic_regression_after_hyperparameter=LogisticRegression(max_iter=1000,solver='liblinear',C=10,penalty='l1')
model_logistic_regression_after_hyperparameter.fit(xtrain,ytrain)
```

```python
# getting the prediction using xtest
predictions=model_logistic_regression_after_hyperparameter.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print("Accuracy of the model: ",accuracy_score(ytest,predictions))
print("The classification report of the model \n",classification_report(ytest,predictions))
```

```python
print("Confusion matric of the model")
sns.heatmap(data=confusion_matrix(ytest,predictions),cmap='winter',annot=True)
```
